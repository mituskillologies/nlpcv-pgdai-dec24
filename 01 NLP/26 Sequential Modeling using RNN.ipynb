```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
list(enumerate(alphabets))
```

```python
int_to_char = dict(enumerate(alphabets))
```

```python
int_to_char
```

```python
int_to_char[5]
```

```python
char_to_int = dict((v, i) for i, v in enumerate(alphabets))
```

```python
char_to_int
```

```python
char_to_int['F']
```

```python
seq_length = 1
x = []
y = []

for i in range(len(alphabets) - seq_length):
    seq_in = [alphabets[i]]
    seq_out = alphabets[i+1]
    print(seq_in,'-->', seq_out)
    x += [[char_to_int[seq_in[0]]]]
    y += [char_to_int[seq_out]]
```

```python
x
```

```python
y
```

##### Prepare the input data

```python
import numpy as np
```

```python
x = np.reshape(x, (25, 1, 1))
```

```python
x
```

```python
x.shape
```

```python
x = x / 25
```

```python
x
```

##### Prepare the output data

```python
y
```

```python
from keras.utils import to_categorical
```

```python
y_new = to_categorical(y)
```

```python
y_new.shape
```

```python
y_new
```

##### Build the model

```python
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN, Input
```

```python
model = Sequential()

model.add(Input((1,1)))
model.add(Dense(25, activation='relu'))
model.add(SimpleRNN(32))
model.add(Dense(26, activation='softmax'))
```

```python
model.summary()
```

```python
from keras.utils import plot_model
```

```python
plot_model(model, show_layer_activations=True, show_layer_names=True,
          show_shapes=True)
```

##### Compile the model

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',
             metrics=['accuracy'])
```

##### Train the model

```python
model.fit(x, y_new, epochs=500, batch_size=1)
```

```python
model.evaluate(x, y_new, batch_size=1)
```

```python
new = 'M'
new = char_to_int[new]
new = np.reshape(new,(1,1,1))
new = new / 25
pred = model.predict(new, verbose=False)
pred = np.argmax(pred)
int_to_char[pred]
```

##### Increasing sequence length

```python
seq_length = 3
x = []
y = []  

for i in range(len(alphabets) - seq_length):
    seq_in = alphabets[i:i + seq_length]
    seq_out = alphabets[i + seq_length]
    x.append([char_to_int[char] for char in seq_in])
    y.append(char_to_int[seq_out])
    print(seq_in,'->',seq_out)
```

```python
x
```

```python
y
```

```python
x_new = np.reshape(x, (23, seq_length, 1))
```

```python
x_new.shape
```

```python
len(y)
```

```python
y_new = to_categorical(y)
y_new.shape
```

```python
x_new = x_new / 25
```

```python
x_new
```

```python
model = Sequential()

model.add(Input((3,1)))
model.add(Dense(25, activation='relu'))
model.add(SimpleRNN(32))
model.add(Dense(26, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',
             metrics=['accuracy'])
```

```python
model.fit(x_new, y_new, epochs=1000, batch_size=1)
```

```python
new = 'HIJ'
val = np.array([[char_to_int[x]] for x in new])
val = val.reshape(1,3,1)
val = val / 25
result = model.predict(val, verbose=False)
result = np.argmax(result)
print(new,'-->',int_to_char[result])
```
