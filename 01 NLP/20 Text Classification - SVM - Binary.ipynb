##### Data Preparation

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag
```

```python
sent = 'Hello friends! How are you? I like Python Programming.'
```

```python
# tokenize the text
tokens1 = word_tokenize(sent)
```

```python
# remove punctuations
tokens2 = [token for token in tokens1 if token.isalpha() or token.isdigit()]
```

```python
tokens2
```

```python
swords = stopwords.words('english')
```

```python
# remove stopwords
tokens3 = [token.lower() for token in tokens2 if token.lower() not in swords]
```

```python
tokens3
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()

# remove the affixes
tokens4 = [ps.stem(token) for token in tokens3]
```

```python
tokens4
```

```python
# Create a function for cleaning text
def clean_text(sent):
    # tokenize the text
    tokens1 = word_tokenize(sent)
    # remove punctuations
    tokens2 = [token for token in tokens1 if token.isalpha()]
    # remove stopwords
    tokens3 = [token.lower() for token in tokens2 if token.lower() not in swords]
    # remove the affixes
    tokens4 = [ps.stem(token) for token in tokens3]
    return tokens4
```

```python
clean_text(sent)
```

```python
sent1 = 'It was nice working with you. Have a great day!'
```

```python
clean_text(sent1)
```

```python
x.apply(clean_text)
```

##### Apply TF*IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(analyzer=clean_text)
```

```python
x_new = tfidf.fit_transform(x)
```

```python
x_new.shape
```

```python
tfidf.get_feature_names_out()
```

```python
len(tfidf.get_feature_names_out())
```

##### Cross-Validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_new, y, random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

##### Build the SVM model

```python
from sklearn.svm import SVC
```

```python
svm = SVC(kernel='linear')
```

```python
svm.fit(x_train, y_train)
```

##### Performance Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
y_pred = svm.predict(x_test)
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
svm = SVC(kernel='rbf')
svm.fit(x_train, y_train)

y_pred = svm.predict(x_test)
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
svm = SVC(kernel='poly')
svm.fit(x_train, y_train)

y_pred = svm.predict(x_test)
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
svm = SVC(kernel='sigmoid')
svm.fit(x_train, y_train)

y_pred = svm.predict(x_test)
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
svm = SVC(kernel='linear')
svm.fit(x_train, y_train)

y_pred = svm.predict(x_test)
print(classification_report(y_test, y_pred))
```
