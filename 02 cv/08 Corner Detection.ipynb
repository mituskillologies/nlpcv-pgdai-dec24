```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/circle.png', 0)
```

```python
plt.imshow(img, cmap = 'gray')
```

```python
help(cv2.cornerHarris)
```

```python
corners = cv2.cornerHarris(img, blockSize=2, ksize=3, k=0.04)
```

```python
corners
```

```python
img[corners > 0.01 * corners.max()] = [255]
```

```python
corners.shape
```

```python
plt.imshow(img)
```

```python
new = corners < corners.max() /2
```

```python
plt.imshow(new, cmap = 'gray')
```
